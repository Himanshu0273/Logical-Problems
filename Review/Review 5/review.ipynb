```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

# Question 1:
Year wise average salary of each job title in the dataset.


```python
file='salaries.csv'
df = pd.read_csv(file,delimiter=',')
avg_per_job=round(df.groupby(['job_title', 'work_year'])['salary'].mean().reset_index(name="Avg_Salary"),2)
years=avg_per_job['work_year'].unique()
for y in range(len(years)):
    plt.subplot(2,3,(y+1))
    data=avg_per_job[avg_per_job['work_year']==years[y]]
    data.set_index('job_title')['Avg_Salary'].plot(kind='bar', title=years[y])
    plt.xticks([])
    plt.legend()
plt.tight_layout()
plt.show()
    
```
