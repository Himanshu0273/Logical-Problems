### Question 11 
Multiple Styles lines in one graph


```python
x = np.array([1,2,3,4,5,6])

y1 = x
y2 = x*2
y3 = x**2

plt.plot(x,y1, 'r--x', x,y2,'g-.|', x,y3,'b:o')

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Multiple Lines with Different styles")
plt.legend(["y=x", "y=2x", "y=x^2"])
plt.grid(color='g', lw=0.5, ls=":")
plt.show()
```

### Question 15
Multiple plots

```python
x= [1,2,3,4,5]
y1=[1,2,3,4,5]
y2=[2,4,6,8,10]
y3=[1,4,9,16,25]
y4=[1,8,27,64,125]

plt.subplot(2,2,1)
plt.plot(x,y1, label="y=x", marker='.')
plt.title("Y=X Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()

plt.subplot(2,2,2)
plt.plot(x,y2, label="y=2x", marker='.')
plt.title("Y=2X Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()


plt.subplot(2,2,3)
plt.plot(x,y3, label="y=x^2", marker='.')
plt.title("Y=X^2 Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()


plt.subplot(2,2,4)
plt.plot(x,y4, label="y=x^3", marker='.')
plt.title("Y=X^3 Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()

plt.suptitle("Line Graphs")
plt.tight_layout()
plt.show()


plt.plot(x,y1, color="r", marker=".", label="y=x")
plt.plot(x,y2, color="b", marker="1", label="y=2x")
plt.plot(x,y3, color="g", marker="+", label="y=x^2")
plt.plot(x,y4, color="k", marker="x", label="y=x^3")
plt.legend()
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Line Graphs")
plt.show()
```
