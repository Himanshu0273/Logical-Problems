## Practice Questions
### Question 1


```python
x=np.array(['Java', 'Python', 'PHP', 'JavaScript', "C#", "C++"])
y=np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])

plt.bar(x,y, label="Popularity")
plt.xlabel("Language")
plt.ylabel("Popularity Rating")
plt.title("Popularity Comparison between all coding Languages")
plt.legend()
plt.show()
```

### Question 3
Use uniform colors for the bars

```python
x=np.array(['Java', 'Python', 'PHP', 'JavaScript', "C#", "C++"])
y=np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
# colors=["red","blue","green","gray","yellow","black"]

plt.bar(x,y, label="Popularity", color="skyblue")
plt.xlabel("Language")
plt.ylabel("Popularity Rating")
plt.title("Popularity Comparison between all coding Languages")
plt.grid(axis='y',ls='--')
plt.legend()
plt.show()
```

### Question 4
Different colors for each bar

```python
x=np.array(['Java', 'Python', 'PHP', 'JavaScript', "C#", "C++"])
y=np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
colors=["red","blue","green","gray","yellow","black"]

plt.bar(x,y, label="Popularity", color=colors)
plt.xlabel("Language")
plt.ylabel("Popularity Rating")
plt.title("Popularity Comparison between all coding Languages")
plt.grid(axis='y',ls='--', lw=0.9)
plt.legend()
plt.show()
```

### Question 5
Attach a text label to each bar(float value of its popularity)

```python
x=np.array(['Java', 'Python', 'PHP', 'JavaScript', "C#", "C++"])
y=np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
# colors=["red","blue","green","gray","yellow","black"]

bars = plt.bar(x,y, label="Popularity", color="cornflowerblue")

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2, height+0.1, f'{height}', ha='center', va='bottom', fontsize=9)
    
plt.xlabel("Language")
plt.ylabel("Popularity Rating")
plt.title("Popularity Comparison between all coding Languages")
plt.grid(axis='y',ls='--')
plt.legend()
plt.tight_layout()
plt.show()
```
