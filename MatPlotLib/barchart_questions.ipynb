### Question 5
Attach a text label to each bar(float value of its popularity)

```python
x=np.array(['Java', 'Python', 'PHP', 'JavaScript', "C#", "C++"])
y=np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
# colors=["red","blue","green","gray","yellow","black"]

bars = plt.bar(x,y, label="Popularity", color="cornflowerblue")

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2, height+0.1, f'{height}', ha='center', va='bottom', fontsize=9)
    
plt.xlabel("Language")
plt.ylabel("Popularity Rating")
plt.title("Popularity Comparison between all coding Languages")
plt.grid(axis='y',ls='--')
plt.legend()
plt.tight_layout()
plt.show()
```

### Question 10
Multiple Graphs together

```python
groups = ['G1', 'G2', 'G3', 'G4', 'G5']
men=[22,30,35,35,26]
women=[25,32,30,35,29]
x=np.arange(len(groups))
width=0.35


plt.bar(x-width/2, men ,width, color="steelblue", label="Men")
plt.bar(x+width/2, women, width, color="salmon", label="Women")
plt.xticks(x, groups)
plt.xlabel("Group Name")
plt.ylabel("Scores")
plt.title("Men vs Women")
plt.legend()
plt.grid(axis='y', ls='--')
plt.show()
```

### Question 
Create bar graph from DataFrame


```python
data={
    'a': [4,2,4,2,2],
    'b': [8,3,7,6,4],
    'c': [5,4,4,4,3],
    'd': [7,2,7,8,3],
    'e': [6,6,8,6,2]    
}

index=[2,4,6,8,10]
df = pd.DataFrame(data, index=index)
# df
df.plot(kind="bar")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Graph from a Data Frame")
plt.legend(title="columns")
plt.tight_layout()
plt.show()
```

### Question 12
Create a bar graph with error bars

```python
mean = [0.2474, 0.1235, 0.1737, 0.1824]
std= [0.3314, 0.2278, 0.2836, 0.2645]

x=np.arange(len(mean))

plt.bar(x, mean, yerr=std, color='steelblue', ec="black", capsize=10)
plt.xlabel("Values")
plt.ylabel("Mean")
plt.title("Mean graph with std deviation")
plt.grid(axis='y', ls='--')
plt.tight_layout()
plt.show()
```

### Question 13
Error bars with name on top

```python
mean = [0.2474, 0.1235, 0.1737, 0.1824]
std= [0.3314, 0.2278, 0.2836, 0.2645]

x=np.arange(len(mean))

bars = plt.bar(x, mean, yerr=std, color='steelblue', ec="black", capsize=10)

for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height+0.01, str(int(mean[i]*100)), ha="center", va="bottom")
plt.xlabel("Values")
plt.ylabel("Mean")
plt.title("Mean graph with std deviation")
plt.grid(axis='y', ls='--')
plt.tight_layout()
plt.show()
```
