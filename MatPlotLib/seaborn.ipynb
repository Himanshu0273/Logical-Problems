# Seaborn

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
dj = sns.load_dataset("dowjones")
crash_df = sns.load_dataset("car_crashes")
iris_df = sns.load_dataset("iris")
tips_df = sns.load_dataset("tips")
```

```python
sns.histplot(crash_df)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

```python
sns.kdeplot(crash_df['alcohol'])
```

```python
sns.pairplot(crash_df)
```

```python
sns.rugplot(crash_df)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, palette="Reds", estimator=np.median)
```

```python
tips_df
```

```python
sns.countplot(x='tip', data=tips_df)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='afmhot')
plt.legend(loc=0)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips_df, hue='sex')
```

```python
flights_df = sns.load_dataset("flights")
flights = flights_df.pivot_table(index='month', columns='year', values='passengers')
plt.figure(figsize=(20,10))
sns.heatmap(flights, cmap="Blues", annot=True)
```
