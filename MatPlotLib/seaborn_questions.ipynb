```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Question 1: Seaborn Bar Plot

```python
url = "https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv"
df = pd.read_csv(url)
sns.barplot(x="sex", y="survived", data=df)#, hue='sex')
# df.head(10)
```

# Question 2: Seaborn Point Plot

```python
url = "https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv"
titanic_df = pd.read_csv(url)
sns.pointplot(x='sex', y='survived',data=titanic_df)
plt.title("Survival Rates on titanic")
plt.show()
```

# Question 3: Seaborn Scatter Plot

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

df=pd.read_csv(url)
sns.scatterplot(x="day", y='total_bill', data=df, hue='sex')
plt.show()
```

# Question 4: Seaborn Violin Plot

```python
tips_df = sns.load_dataset("tips")
sns.violinplot(x='sex', y='total_bill', data=tips_df, hue='sex')
```

# Question 5: Seaborn Violin Plot 2

```python
df = sns.load_dataset("iris")
sns.violinplot(x='species', y='sepal_length', data=df)
plt.show()
```

# Question 6: Seaborn Box Plot 1

```python
url = 'https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv'

df = pd.read_csv(url)
df.head()

sns.boxplot(x='continent', y='lifeExp', data=df)
plt.title("Life Expectancy vs Continent")
plt.xlabel("Continent")
plt.ylabel("Life Expectancy")
plt.show()
```

# Question 7: Seaborn Box Plot 2

```python
tips_df = sns.load_dataset("tips")

sns.boxplot(x='day', y='tip', data=tips_df)
plt.title("Tip by day")
plt.show()
```

# Question 8: Seaborn Swarm Plot 1


```python
tips_df = sns.load_dataset("tips")
# tips_df.head()

sns.swarmplot(x='size', y='total_bill', data=tips_df)
plt.title("Total bill vs size")
plt.show()
```

# Question 9: Seaborn Swarm Plot 2

```python
sns.swarmplot(x='day', y="total_bill", data=tips_df, hue="sex")
plt.title("Total Bill vs day")
plt.show()
```
