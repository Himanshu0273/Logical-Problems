# Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
import os
```

```python
y=np.array([35,25,25,15])
mylabels=np.array(['A', 'B', 'C', 'D'])
mycolors=np.array(['r','b','g','y'])
myexplode=np.array([0.1,0,0.3,0])


plt.pie(y, labels=mylabels, startangle=90, colors=mycolors, explode=myexplode, shadow=True)
plt.show()
```

### Question 1
Plot pie chart for the languages

```python
ratings = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
lang_name = np.array(['Java','Python','PHP','JS','C++','C'])
plt.pie(ratings, labels=lang_name, shadow=True)
plt.show()
```

### Question 2
Title the pie chart as "Languages Popularity Distribution"

```python
ratings = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
lang_name = np.array(['Java','Python','PHP','JS','C++','C'])
plt.pie(ratings, labels=lang_name, shadow=True)
plt.legend(title="Language")
plt.title("Popularity of the programming language")
plt.show()
```

### Question 3
Read from a csv file

```python
countries=[]
medals=[]
with open(r"data.csv", 'r') as input_file:
    reader = csv.DictReader(input_file)
    for row in reader:
        countries.append(row["country"])
        medals.append(row["medals"])
        
plt.figure(figsize=(8,8))
plt.pie(medals, labels=countries)
plt.title("Gold Medals for each country")
plt.axis("equal")
plt.show()
```
